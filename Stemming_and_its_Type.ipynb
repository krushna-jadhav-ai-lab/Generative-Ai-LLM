## Stemming 

Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma.
Stemming is important in natural language understanding (NLU) and natural language processing(NLP)

```python
### Sentiment  or classification problem 
### comments on the product is a positive review or negative review
## Reviews ---> for ex - eating,eat,eaten --- > eat is common
```

```python
practice_words = [
    "eat", "eats", "eating", "eaten", "ate",
    "run", "runs", "running", "ran",
    "walk", "walks", "walking", "walked",
    "study", "studies", "studying", "studied",
    "fly", "flies", "flying", "flew",
    "goose", "geese", "mouse", "mice", "foot", "feet",
    "beauty", "beautiful", "beautify", "beautifully",
    "happy", "happier", "happiest", "happily",
    "child", "children", "childhood",
    "organize", "organizes", "organizing", "organization"
]

from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
for word in practice_words:
    print(word+"--->"+stemming.stem(word))
```

```python
stemming.stem("congratulations")  ### Major disadvantage of stemming
```

```python
stemming.stem("sitting")
```

### RegexStemmer class

NLTK has RegexStemmer class with the help of which we can easily implement Regular Expression stemmer algorithms. It basically takes a single expression and removes any prefix or suffix that matches the expression.Let us see an examples

```python
from nltk.stem import RegexpStemlmer
```

```python
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem("eating")
```

```python
reg_stemmer.stem("ingeating")
```

### Snowball stemmer


```python
from nltk.stem import SnowballStemmer
```

```python
snowballstemmer=SnowballStemmer("english")
```

```python
for word in practice_words:
    print(word+"--->"+snowballstemmer.stem(word))
```

```python
stemming.stem("fairly"),stemming.stem("sportingly")
```
