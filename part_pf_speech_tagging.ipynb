```python
speech = """
I'm standing here today, thinking about how far we've come. 
Dr. Smith once said, 'The journey of a thousand miles begins with a single step.' 
But what if that step is taken on rocky ground? 
The children are playing outside, their feet hitting the pavement, 
while the geese fly overhead in beautiful patterns. 
Yesterday, we ate lunch while it was raining; today, we're eating in the sun. 
Technology is organizing our lives, but are we truly organized? 
It doesn't matter if you're a mouse or a lion; everyone has a role!
Visit us at https://nlp-practice.com to learn more.
"""
```

Tag	Description	Example
NN	Noun, singular	dog, desk, book
NNS	Noun, plural	dogs, desks
VB	Verb, base form	eat, go
VBG	Verb, gerund/present participle	eating, going
VBD	Verb, past tense	ate, went
JJ	Adjective	quick, brown, lazy
RB	Adverb	quickly, very
PRP	Personal Pronoun	I, he, she

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
stemmer=PorterStemmer()
sentences=nltk.sent_tokenize(speech)
nltk.download('averaged_perceptron_tagger_eng')
```

```python
# we will find out the pos tag

for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[word for word in words if word not in set(stopwords.words('english'))]
    pos_tag=nltk.pos_tag(words)
    print(pos_tag)
    #sentences[i]=' '.join(words) # conerveting all list of the wprds in sentences
```
