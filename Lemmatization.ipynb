## WordNet Lemmatizer
The WordNet Lemmatizer is a tool in NLTK that links words to their actual dictionary definitions (from the WordNet database) to find their "base" or "dictionary" form, which is called a lemma.Unlike a Stemmer, which just chops off the ends of words (like "eating" $\rightarrow$ "eat"), a Lemmatizer is smarterâ€”it understands grammar and context.The Key DifferenceStemming: Uses rules to slice words. It might turn "studies" into "studi".Lemmatization: Uses a dictionary. It turns "studies" into "study" and "mice" into "mouse".

```python
from nltk.stem import WordNetLemmatizer
```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
'''
postag - Noun - n
verb - v
adjective- a
adverb - r
'''
lemmatizer.lemmatize("going", pos='v')
```

```python
practice_words = [
    "eat", "eats", "eating", "eaten", "ate",
    "run", "runs", "running", "ran",
    "walk", "walks", "walking", "walked",
    "study", "studies", "studying", "studied",
    "fly", "flies", "flying", "flew",
    "goose", "geese", "mouse", "mice", "foot", "feet",
    "beauty", "beautiful", "beautify", "beautifully",
    "happy", "happier", "happiest", "happily",
    "child", "children", "childhood",
    "organize", "organizes", "organizing", "organization"
]
```

```python
for word in practice_words:
    print(word+"-->"+lemmatizer.lemmatize(word, pos='v'))
```
